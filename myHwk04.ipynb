The solution to this is to use the `random()` function to generate a random

and then successively add the values in `p` until you equal or exceed `r`. 
You are effectively dividing the line $[0..1)$ into regions corresponding
to the various probabilities (length == probability). 

For example, if you have X = [0,1,3] and p = [0.1,0.3,0.6], then you can diagram
the probabilities as follows:





- If you generate r = 0.26 then you would return 1, because as you scan from the left, r > 0.1, but r <= 0.1 + 0.3 = 0.4;  

- If you generate r = 0.679 then you would return 3, because as you scan from the left, r > 0.1, and r > 0.1 + 0.3 = 0.4, but r <= 0.1 + 0.3 + 0.6 = 1.0. 

The effect of this is that you are selecting a random point in the interval, and finding out which region it lands in, where the
length of each region is the probability. 

```python
# Choose a single element of X according to p = None (equiprobably) or p = list of probs
# You may assume that len(X) = len(p)

def choose1(X,p=None):
        if p == None: 
            return choice(X)
        else:
            val = [ 1 * p[0] ]
            for i in range(1,len(p)):
                holder = (1 * p[i]) + val[i-1]
                val += [holder]
            
            r = random()
            for i in range(len(val)):
                if i != 0:
                    if val[i-1] <= r <= val[i]:
                        return X[i]
                elif i == 0:
                    if 0 <= r <= val[0]:
                        return X[0]
            
# Now implement the main function, using choose1(...)

def my_choice(X,size=None,p=None):
        if size == None:
            return choose1(X,p)                         # Your code here
        else:
            ans = [ choose1(X,p) for i in range(size)]
            return ans

# test it




seed(0)

X = [1,2,3,4,5,6]

print( choose1(X), choose1(X), choose1(X), choose1(X) )  # should print 5 6 1 4

p = [0.1,0.1,0.1,0.2,0.2,0.3]
print( choose1(X,p), choose1(X,p), choose1(X,p), choose1(X,p) )  # should print 6 6 5 4
```
